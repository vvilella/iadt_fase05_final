```python
from ultralytics import YOLO
from pathlib import Path

# repo root (notebooks/ -> ../)
REPO_ROOT = Path.cwd().parent

DATASET_YAML = REPO_ROOT / "data" / "yolo" / "dataset.yaml"

print("Dataset YAML exists:", DATASET_YAML.exists())

model = YOLO("yolov8n.pt")

results = model.train(
    data=str(DATASET_YAML),
    epochs=20,
    imgsz=640,
    batch=8,
    project=str(REPO_ROOT / "runs"),
    name="detect_train_mvp",
)

print("Done.")
```
